# Animal Tree

```python
class Tree:
    def __init__(self, cargo, left = None, right = None, parent = None):
        self.cargo = cargo
        self.left = left
        self.right = right
        self.parent = parent
        
    def getCargo(self):
        return self.cargo
    
    def getLeft(self):
        return self.left
    
    def getRight(self):
        return self.right
    
    def setLeft(self, tree):
        self.left = tree
        
    def setRight(self, tree):
        self.right = tree
        
    def setCargo(self, cargo):
        self.cargo = cargo
```

```python
# Now using preorder storing and traversing
def treeToTextFile(tree,textfile):
    texttree = treeToString(tree).strip('$')# to remove extra spaces from left and right
    with open(textfile,'w') as wf:
        wf.write(texttree)

def treeToString(tree):
    if tree == None: return ""
    return tree.cargo + '$' + treeToString(tree.left) + '$' + treeToString(tree.right) + '$'  # breaking cargos by $
```

```python
# Now using preorder traversing 
def textFileToTree(textlist):
    # deviding into 3 parts to read left right and cargo
    i = 0
    parent = Tree('',None,None)#None , Dummy tree
    flag = False
    head = None
    l = len(textlist)
    while i<l:#not textlist[i][0]=='*':
        cargo = textlist[i]
        tree = Tree(cargo)
        if i == 0: head = tree

        if not flag:
            parent.left = tree # traversing left
            #print(cargo,'in the not flag')
            tree.parent = parent
            parent = tree
            
        if flag:
            #print(cargo,'in the flag',parent.cargo,'<- parent')
            parent.right = tree
            tree.parent = parent
            parent = parent.parent
            
            
        if cargo[0] == '*' and not flag:
            #print(cargo,'in the *')
            parent = tree.parent
            flag = True
        
        if cargo[0] != '*' and flag:
            #print(cargo,'in the not *')
            parent = tree
            flag = False
        i+=1
    return head
    
    
def textFiletoList(textfile):
    textlist=[]
    with open(textfile, 'r') as rf:
        textlist = list(filter(None,rf.readline().split('$')))
    return textlist
```

# main self updating animal it can tell which animal you are thinking off

```python
def animal():
    # start with a singleton
    #root = Tree('*bird') # Marking with '*'
    # After the knowledge gained
    textlist = textFiletoList('animal_data.txt')
    root = textFileToTree(textlist)
    
    while True:
        print()
        if not yes("Are you thinking of an animal? "):
            treeToTextFile(root,'animal_data.txt')# before exiting storing the gained knowledge in the animal_data.txt
            break
            
        tree = root
        while tree.getLeft() != None:
            prompt = tree.getCargo()+'?'
            if yes(prompt):
                tree = tree.getRight()
            else:
                tree = tree.getLeft()
            
        guess = tree.getCargo()
        prompt = "Is it a "+ guess.lstrip('*') + "? "
        if yes(prompt):
            print("I rule! hahahahaha")
            continue
            
        # get new information 
        prompt = "What is the animal's name? "
        animal = input(prompt)
        prompt = "What question would distinguish a %s from a %s "
        question = input(prompt % (animal,guess.lstrip('*')))
        
        # add new information to the tree
        tree.setCargo(question)
        prompt = "If the animal were %s the answer would be ? "
        animal = '*'+animal # marking the new animal's name with '*'
        if yes(prompt % animal.lstrip('*')):
            tree.setLeft(Tree(guess))
            tree.setRight(Tree(animal))
        else:
            tree.setLeft(Tree(animal))
            tree.setRight(Tree(guess))
            
def yes(ques):
    #from string import lower
    ans = input(ques).lower()#lower(input(ques))
    return (ans[0] == 'y')
```
